## For Else 

```python
for i in l:
    print(i)
else:
    print("If for loop is going to complete execution then it will come here")
```

## Break statement

```python
for i in l:
    if i == 'Name':
        break
    print(i)
else:
    print("Check this statement")
```

```python
s = 'Astajyoti'
for i in s:
    if i == 'k':
        break
else:
    print('dont execute it unless and untill it is not printing my name')
```

# While Loop

```python
a = 1
while a<6:
    print(a)
    a = a+1
```

```python
a = 1
while a <8:
    print(a)
    if a ==4:
        break
    a= a+1
```

### Continue Statement

```python
a = 1
while a<5:
    print(a)
    if a == 3:
        continue
    a = a+1
```

--> It will go in a infinite loop. Whn a = 3, it will take to While statement withut adding 1 and it will continue in infinite loop
